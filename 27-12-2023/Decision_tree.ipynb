```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
 from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
x = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
x
```

```python
iris.target
```

```python
x["Species"] = iris.target 
```

```python
x
```

```python
y = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
y
```

```python
y = pd.get_dummies(y) # this is one hot encoding variable, it is use to convert categorical data into binary form
```

```python
y
```

```python
x.columns
```

```python
x = x[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape, x_test.shape)
```

```python
print(y_train.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x_train, y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
y_predict
```

```python
fig = plt.figure(figsize=(30,30))
from sklearn import tree
tree.plot_tree(clf,filled=True)
```

```python
prediction = np.array(y_predict).argmax(axis=1)
```

```python
prediction
```

```python
species = np.array(y_test).argmax(axis=1)
```

```python
species
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(species, prediction))
```

```python
print(confusion_matrix(species, prediction))
```
